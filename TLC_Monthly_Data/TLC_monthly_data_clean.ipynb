# TLC monthly Data

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/Users/adam/Bootcamp_API_project/Data/data_reports_monthly_indicators.csv")
data.head()
```

```python
data = data[["Month/Year","License Class","Trips Per Day","Vehicles Per Day","Avg Hours Per Day Per Driver",
            "Avg Minutes Per Trip"]]
data.head()
```

```python
data["year"] = [data["Month/Year"][x].split("-")[0] for x in range(len(data["Month/Year"]))]
data["month"] = [data["Month/Year"][x].split("-")[1] for x in range(len(data["Month/Year"]))]
data.head()
```

```python
category = []
for x in range(len(data["Month/Year"])):
    word = data["License Class"][x].split("-")[0].strip()
    
    if word == "FHV":
        category.append(word)
    else :
        category.append("Taxi")
data["Category"] = category
data.head()
```

```python
new_data = data.copy()
new_data.dtypes
```

```python
new_data["Trips Per Day"] = new_data['Trips Per Day'].str.replace(',','').astype(int)
new_data["Vehicles Per Day"] = new_data['Vehicles Per Day'].str.replace(',','').astype(int)
new_data = new_data.convert_objects(convert_numeric=True)
new_data.dtypes
```

```python
TLC_monthly_data = new_data.groupby(["year","month","Category"]).agg({"Trips Per Day":"sum",
                                              "Vehicles Per Day":"sum",
                                              "Avg Hours Per Day Per Driver":"mean",
                                              "Avg Minutes Per Trip":"mean"})
TLC_monthly_data
```

```python
TLC_monthly_data.to_csv("TLC_Monthly_Data.csv")
```
